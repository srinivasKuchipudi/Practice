### Bollinger Bands calculation
Calculate all the bollinger bands here

```python
bb = ta.BBANDS(df[field].values,matype=ta.MA_Type.SMA)
df["BBUpper"] = bb[0]
df["BBMiddle"] = bb[1]
df["BBLower"] = bb[2]
```

```python
df["LONGEMA"]= ta.EMA(df["CLOSE"].values, timeperiod=200)
df["MEDIUMEMA"]= ta.EMA(df["CLOSE"].values, timeperiod=50)
df["SHORTEMA"]= ta.EMA(df["CLOSE"].values, timeperiod=7)
```

```python
df.round({"Open":1, "TOTTRDQTY":0, "BBUpper":1, "BBMiddle":1, "BBLower":1, "SHORTEMA":1 })
df.dtypes
```

```python
df["trendangle"]=ta.LINEARREG_ANGLE (df[field].values, timeperiod=7)
df['trend'] = df.apply(lambda x: 'U' if x['trendangle']>0 else 'D', axis=1)
```
